2. Let's disect the sentence "it is important to note that outcomes of tests refer to the population parameter, rather than the sample statistic! As such, the result that we get is for the population." It is important to understand that even though we only perform tests on a small portion of the population, often called the sample, we want to make inferences about the characteristics of the entire population. So, the outcomes of our tests refer to the entire population, instead of the sample we took. The rest of this question can be best explained with an example. Suppose I want to check that the average income of a family in my neighbourhood is 50,000. I can then set up a null hypothesis (H0), which is essentially my default hypothesis. It is a claim which I can test, but if I don't find anything significant then I will just have to assume my null hypothesis as the truth. It usually represents the status quo. So if I already believe that the average income in my neighbourhood is 50,000, then my null hypothesis will be that the average income in my neighbourhood is 50,000. This is often written as H0 : μ = μ0. μ is the mean of the sample, and μ0 is the value that we assume for the mean in the null hypothesis. Then, I can set up my alternative hypothesis, or HA, which would be used when H0, my null hypothesis is false. Now how do I decide whether to accept or reject the null hypothesis? I have to collect sample values, (represented as xi) from my population. At this stage, I might go to 10 houses out of the 1000 in my neighbourhood, and I would ask them for their income. Each income I get would be a sample value, and when I take the mean of these values, I get the sample mean. This would then be used in my hypothesis testing, which is to determine whether to accept or reject the null hypothesis. If the null hypothesis is rejected (ie there is an observable difference), then the alternative hypothesis will be considered stronger than the null. If the null hypothesis is rejected, then we can use the alternate hypothesis for the population. 

3. We want to imagine that it is true, but the reason we do this is to consider the likelihood that it is true. If it is very likely that it is true, then we will have a very large p value. If it is unlikely, then we will have a much smaller p value.

4. The p-value in essence represents the probability that we would achieve that test statistic, while assuming that the null hypothesis is true. The test statistic is based on the actual tests we did (it could be mean, median, etc), while the null hypothesis is what represents no significant change, or nothing happened. When we calculate a p-value and find it to be very small (often 0.05 is used as a benchmark), that tells us it is very far away from the sampling distribution, which is the theoretical distribution of the test statistic under the assumption of the null hypothesis. So, it is pretty unlikely to have occured, and makes our assumption (the null hypothesis) look pretty ridiculous. Conversely, if we get a very large p-value that tells us that the test statistic is similar to the sampling distribution (theoretical distribution), so it seems pretty reasonable to have gotten the test statistic given the sampling distribution.

```python
# 5
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_couples = 124    # number of couples
observed_right_tilts = 80  # observed number of couples tilting right
p_null = 0.5       # null hypothesis probability for right tilt
n_simulations = 10000  # number of simulations

# Simulate the number of right tilts under the null hypothesis (50/50)
simulated_tilts = np.random.binomial(n=n_couples, p=p_null, size=n_simulations)

# Calculate the p-value (proportion of simulations where number of right tilts >= observed number)
p_value = np.sum(simulated_tilts >= observed_right_tilts) / n_simulations

# Output the p-value
print(f"P-value: {p_value:.4f}")

# Plotting the simulation results
plt.hist(simulated_tilts, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(observed_right_tilts, color='red', linestyle='dashed', linewidth=2)
plt.title(f'Simulation of Right Tilts (n={n_simulations} simulations)\nRed line = Observed right tilts = {observed_right_tilts}')
plt.xlabel('Number of right tilts in a sample of 124 couples')
plt.ylabel('Frequency')
plt.show()
```

8. This assignment involves determining whether students in a STA130 class are able to correctly tell whether milk or tea is poured first when preparing a cup of tea. The sample size for this experiment is 80 students out of the entire group. It bears similarity to Fisher's tea experiment with Bristol, in which Fisher prepared Bristol 8 cups of tea, 4 with milk first and 4 with tea first in an attempt to determine whether or not Bristol could actually tell which was poured first. Bristol identified them all correctly, but Fisher wondered if Bristol had only happened to guess them all correctly. A similar problem applies to this experiment, what if the 80 students in STA130 only guessed, and more of them happened to guess correctly? We will perform statistical testing to determine whether STA130 students can correctly identify whether tea or milk was poured first. In this case, the null hypothesis (H0) will be that 50% or less of the population would be able to correctly identify the difference. This is selected as if students guessed, then their guess would be a 50/50 shot, so theoretically 50% of them would guess correctly and 50% incorrectly. This also means that we will be performing a right-tailed test, as we only want to test if the number of students correctly identifying is greater than 0.5. The alternative hypothesis (HA) will then be that more than 50% of the population would be able to correctly identify whether milk or tea was poured first. 

```python
import numpy as np
import statsmodels.api as sm
import plotly.graph_objects as go

# Make simulation reproducible
np.random.seed(1)

# Given data
n_students = 80
n_correct = 49
p_null = 0.5  # Null hypothesis proportion
number_of_simulations = 10000  

# Initialize array to store simulated proportions under H0
simulated_proportions = np.zeros(number_of_simulations)

# Perform the simulations
for i in range(number_of_simulations):
    # Generate random outcomes assuming H0 is true
    random_outcomes = np.random.choice([0, 1], size=n_students, p=[1-p_null, p_null])
    # the above line randomly selects one out of two answers, 0 as incorrect and 1 as correct
    # the p of each of these is calculated from p_null, and I had told the chatbot that it was supposed to be 0.5
    # so the chances of correct and incorrect are 0.5 each (equal to each other)
    
    # Calculate the proportion of correct answers in the simulation
    simulated_proportions[i] = random_outcomes.mean()

# Calculate sample proportion
p_sample = n_correct / n_students

# Perform the z-test for proportion (right-tailed test)
z_stat, p_value = sm.stats.proportions_ztest(n_correct, n_students, p_null, alternative='larger')
# z_stat performs the z-test for proportions, alternative = 'larger' specifies that this is a right-tailed test
# Calculate the confidence interval for the sample proportion
conf_int = sm.stats.proportion_confint(n_correct, n_students, alpha=0.05, method='normal')
# confidence interval specified as 95% as alpha=0.05

# all the below code is for making a figure to represent the data  
# Output results
print(f"Sample Proportion: {p_sample:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value (right-tailed): {p_value:.4f}")
print(f"95% Confidence Interval: [{conf_int[0]:.4f}, {conf_int[1]:.4f}]")


# Create a bar plot for sample proportion
fig = go.Figure()

# Add bars for the sample proportion and the null hypothesis proportion
fig.add_trace(go.Bar(
    x=['Sample Proportion', 'Null Hypothesis Proportion'],
    y=[p_sample, p_null],
    text=[f'{p_sample:.2f}', f'{p_null:.2f}'],
    textposition='auto',
    marker_color=['blue', 'orange']
))

# Add confidence interval as a line
fig.add_trace(go.Scatter(
    x=['Confidence Interval Lower', 'Confidence Interval Upper'],
    y=[conf_int[0], conf_int[1]],
    mode='lines+text',
    line=dict(color='green', dash='dash'),
    text=[f'{conf_int[0]:.2f}', f'{conf_int[1]:.2f}'],
    textposition='top center',
    name='95% Confidence Interval'
))

# Update layout
fig.update_layout(
    title='Proportion of Correct Answers vs Null Hypothesis Proportion (Right-Tailed Test)',
    yaxis_title='Proportion',
    yaxis=dict(tickvals=np.arange(0, 1.1, 0.1)),
    showlegend=False
)

# Show the figure
fig.show()
```

In the above code, we have run a simulation to determine what the null hypothesis proportion should be. Then, we ran statistical tests comparing the sample proportion to the null hypothesis proportion in order to determine the p-value and confidence interval of the sample statistic to make an inference on the population. We have now a theoretical proportion of students answering correcctly (if the null hypothesis were true), and the actual sample proportion from students in the STA130 class. The above visualization shows the sample proportion compared to the null hypothesis proportion, as well as a figure of the 95% confidence interval (0.51, 0.72). Additionally, the p value was 0.0194 which is less than 0.05. Recall that we performed a right tailed test, which means that our null hypothesis (H0) was that 50% or less of the population would be able to correctly identify whether milk or tea was poured first. Since the p value is less than 0.05, and our confidence interval (0.51, 0.72) does not include 0.5, we can reject the null hypothesis in favor of the alternative hypothesis, that more than 50% of the population can correctly identify which was poured first. Since we infer that more than 50% of the population can correctly identify which was poured first, we can infer that students are not simply guessing and getting the right answer, and that some of them are actually able to tell which was poured first.
